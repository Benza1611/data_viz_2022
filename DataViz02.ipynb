# Import Data

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df.head()
```

# Scatter Plot

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter([1,-1],[2,-1])
```

```python
plt.scatter(df['PetalLength'],df['SepalLength'])
```

## เปลี่ยนสี

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c='r')
```

## ระบุสีให้แต่ละจุด

```python
set(df['Name'])
```

### แทนค่าชื่อดอกไม้ด้วยตัวอักษรแทนสี (r g b c m y k w)

```python
df2 = df.replace({'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'})
```

```python
df2.head()
```

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c=df2['Name'])
```

## เพิ่มรายละเอียดกราฟ

### ชื่อแกน

```python
plt.scatter(df['PetalLength'],df['SepalLength'],c=df2['Name']) #วาดกราฟ
plt.xlabel('PetalLength') #ชื่อแกนX
plt.ylabel('SepalLength') #ชื่อแกนy
```

### ชื่อ marker

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50],label='setosa') #วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100],label='versicolor') #วาดกราฟ
plt.scatter(df2['PetalLength'][100:150],df2['SepalLength'][100:150],c=df2['Name'][100:150],label='verginiga') #วาดกราฟ
plt.legend() #อธิบาย marker
plt.xlabel('PetalLength') #ชื่อแกนX
plt.ylabel('SepalLength') #ชื่อแกนy
```

### ชื่อกราฟ

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50],label='setosa') #วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100],label='versicolor') #วาดกราฟ
plt.scatter(df2['PetalLength'][100:150],df2['SepalLength'][100:150],c=df2['Name'][100:150],label='verginiga') #วาดกราฟ
plt.legend() #อธิบาย marker
plt.xlabel('PetalLength') #ชื่อแกนX
plt.ylabel('SepalLength') #ชื่อแกนy
plt.title('comparing petal&sepal length of different iris flowers'); #ชื่อกราฟ
```

### ใช้ขนาดของ marker ในการแสดงค่าใน column

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50],label='setosa',s=df2['PetalWidth'][:50]) #วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100],label='versicolor',s=df2['PetalWidth'][50:100]) #วาดกราฟ
plt.scatter(df2['PetalLength'][100:150],df2['SepalLength'][100:150],c=df2['Name'][100:150],label='verginiga',s=df2['PetalWidth'][100:150]) #วาดกราฟ
plt.legend() #อธิบาย marker
plt.xlabel('PetalLength') #ชื่อแกนX
plt.ylabel('SepalLength') #ชื่อแกนy
plt.title('comparing petal&sepal length of different iris flowers'); #ชื่อกราฟ
```

### ขยายขนาด marker ที่เล็กเกินไป

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],c=df2['Name'][:50],label='setosa',s=20*df2['PetalWidth'][:50]) #วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],c=df2['Name'][50:100],label='versicolor',s=20*df2['PetalWidth'][50:100]) #วาดกราฟ
plt.scatter(df2['PetalLength'][100:150],df2['SepalLength'][100:150],c=df2['Name'][100:150],label='verginiga',s=20*df2['PetalWidth'][100:150]) #วาดกราฟ
plt.legend() #อธิบาย marker
plt.xlabel('PetalLength') #ชื่อแกนX
plt.ylabel('SepalLength') #ชื่อแกนy
plt.title('comparing petal&sepal length of different iris flowers'); #ชื่อกราฟ
```

### ปรับความโปรงใส่ของ marker เพื่อให้มองเห็นจุดที่ถูกบัง (Buble Chart)

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],alpha=0.5,c=df2['Name'][:50],label='setosa',s=20*df2['PetalWidth'][:50]) #วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],alpha=0.5,c=df2['Name'][50:100],label='versicolor',s=20*df2['PetalWidth'][50:100]) #วาดกราฟ
plt.scatter(df2['PetalLength'][100:150],df2['SepalLength'][100:150],alpha=0.5,c=df2['Name'][100:150],label='verginiga',s=20*df2['PetalWidth'][100:150]) #วาดกราฟ
plt.legend() #อธิบาย marker
plt.xlabel('PetalLength') #ชื่อแกนX
plt.ylabel('SepalLength') #ชื่อแกนy
plt.title('comparing petal&sepal length of different iris flowers'); #ชื่อกราฟ
```

### เปลี่ยนหน้าตาของ marker เพื่อความสวยงาม

```python
plt.scatter(df2['PetalLength'][:50],df2['SepalLength'][:50],alpha=0.5,c=df2['Name'][:50],label='setosa',s=20*df2['PetalWidth'][:50],marker='$S$') #วาดกราฟ
plt.scatter(df2['PetalLength'][50:100],df2['SepalLength'][50:100],alpha=0.5,c=df2['Name'][50:100],label='versicolor',s=20*df2['PetalWidth'][50:100],marker='X') #วาดกราฟ
plt.scatter(df2['PetalLength'][100:150],df2['SepalLength'][100:150],alpha=0.5,c=df2['Name'][100:150],label='verginiga',s=20*df2['PetalWidth'][100:150],marker='^') #วาดกราฟ
plt.legend() #อธิบาย marker
plt.xlabel('PetalLength') #ชื่อแกนX
plt.ylabel('SepalLength') #ชื่อแกนy
plt.title('comparing petal&sepal length of different iris flowers'); #ชื่อกราฟ
```
