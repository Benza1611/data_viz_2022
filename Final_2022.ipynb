# Data Preprocessing

```python
data = data1[['user','RowKey','date','ov','phone','site','time','urine','PartitionKey','title','age']]
data
```

```python
A=list() 
for each_row in data.iterrows(): # loop each_row ใน data
    if len(str(each_row[1]['user']))== 5 : #เลือกตัวอักษรที่เท่ากับ 5
        if str(each_row[1]['user']) <= '99999' : #ถ้า user  น้อยกว่า 100000
            A.append(True) #ให้เพิ่มเข้าไปใน A
        elif str(each_row[1]['user']) >= '99999' : #ถ้า user มากกว่า 100000
            A.append(False) #ไม่ให้เพิ่มเข้าไปใน A
    else:
        A.append(False) #ถ้าไม่ใช่ทั้งหมด ไม่ให้เพิ่มเข้าไปใน A

data_test = data[A] # ใส่ตัวแปร
data_test# เรียกดูตาราง
```

```python
set(data_test['user']) 
```

```python
B=list() 
for each_row in data_test.iterrows(): # loop each_row ใน data_test
    if len(str(each_row[1]['site']))== 5 : #เลือกตัวอักษรที่เท่ากับ 5
        if str(each_row[1]['site']) <= '99999' : #ถ้า user  น้อยกว่า 100000
            B.append(True) #ให้เพิ่มเข้าไปใน B
        elif str(each_row[1]['site']) >= '99999' : #ถ้า user มากกว่า 100000
            B.append(False) #ไม่ให้เพิ่มเข้าไปใน B
    else:
        B.append(False) #ถ้าไม่ใช่ทั้งหมด ไม่ให้เพิ่มเข้าไปใน B

data_test2 = data_test[B] # ใส่ตัวแปร
data_test2# เรียกดูตาราง
```

```python
set(data_test2['site']) 
```

```python
data_test2 = data_test2[data_test2['site']!= ' 0502']
```

```python
data_test2 = data_test2[(data_test2['PartitionKey'].astype(str).str.isdigit()) & (data_test2['PartitionKey'].str.len() == 5)]      # เลือกมาเฉพาะ คอลัมน์ PartitionKey ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
print(data_test2.shape)                                                                            
data_test2
```

## แทนค่าข้อมูลที่ผิดปกติ

```python
data_test2[['ov']] = data_test2[['ov']].fillna('0') ##แทนค่า NaN  เป็น 0
data_test2[['time']] = data_test2[['time']].fillna('ไม่ระบุ') ##แทนค่า NaN  เป็น ไม่ระบุ
data_test2[['urine']] = data_test2[['urine']].fillna('ไม่ระบุ') ##แทนค่า NaN  เป็น ไม่ระบุ
data_test2[['title']] = data_test2[['title']].fillna('ไม่ระบุ') ##แทนค่า NaN  เป็น ไม่ระบุ
data_test2[['age']] = data_test2[['age']].fillna('ไม่ระบุ') ##แทนค่า NaN  เป็น ไม่ระบุ
```
