```python
data_covid2 = pd.read_csv('/content/drive/My Drive/project2021 data/confirmed-cases-since-120864.csv') #เข้าไปที่ my drive -> โฟลเดอร์ project2021 data -> confirmed-cases-since-120864.csv
data_covid2
```

##ข้อมูลผู้ป่วย 26/10/2021 - 27/02/2022


```python
data_covid3 = pd.read_csv('/content/drive/My Drive/project2021 data/confirmed-cases-since-271064.csv') #เข้าไปที่ my drive -> โฟลเดอร์ project2021 data -> confirmed-cases-since-271064.csv
data_covid3 = data_covid3.drop('Unnamed: 11', axis=1)
data_covid3
```

##ข้อมูลผู้ป่วย 28/02/2022 - 18/03/2022

```python
data_covid4 = pd.read_csv('/content/drive/My Drive/project2021 data/confirmed-cases-since-280265.csv') #เข้าไปที่ my drive -> โฟลเดอร์ project2021 data -> confirmed-cases-since-280265.csv
data_covid4
```

```python
data_covid = pd.concat([data_covid1,data_covid2,data_covid3,data_covid4])
data_covid
```

###ดูค่า Missing ของ data_covid

```python
data_covid.isnull().any() ##ดูค่า Missing ในตาราง data_covid
```

### กำจัดค่า Missing 

```python
set(data_covid['province_of_isolation']) #ตรวจสอบค่าใน province_of_isolation
```

```python
data_covid['province_of_isolation'].replace({'กทม':'กรุงเทพมหานคร','ขอนแก่น\xa0':'ขอนแก่น','ภููเก็ต':'ภูเก็ต'}, inplace=True ) #แทนที่ข้อมูลที่ผิดปกติ
```

```python
data_covid[['Unit']] = data_covid[['Unit']].fillna('ปี') ##แทนค่า NaN ใน Unit เป็น ปี
data_covid = data_covid.fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
```

```python
data_covid.isnull().any() ##ดูค่า Missing ในตาราง data_covid
```
