###จัดกลุ่มข้อมูลตามวัน

```python
A0 = covid[['announce_date','No.']].groupby(['announce_date']).count()
A0.rename(columns = {'No.':'จำนวนผู้ติดเชื้อ'}, inplace = True)
A0
```

```python
L = covid[covid['ชาย'] != False ]
A1 =  L[['announce_date','sex','ชาย']].groupby(['announce_date','ชาย']).count()
A1.rename(columns = {'sex':'ชาย'}, inplace = True)
A1
```

```python
L2 = covid[covid['หญิง'] != False ]
A2 =  L2[['announce_date','sex','หญิง']].groupby(['announce_date','หญิง']).count()
A2.rename(columns = {'sex':'หญิง'}, inplace = True)
A2
```

```python
A3 = covid[['announce_date','province_of_isolation','จำนวนคนกักตัวในจังหวัด']].groupby(['province_of_isolation']).mean()
A3
```

```python
A4 =  covid[['announce_date','province_of_isolation','จังหวัดที่กักตัว']].groupby(['province_of_isolation']).min()
A4
```

```python
A5 = A4.merge(A3,how='left',left_on='จังหวัดที่กักตัว',right_on='province_of_isolation')
A5
```

```python
A5.info() ##ดูประเภทของข้อมูล
```
